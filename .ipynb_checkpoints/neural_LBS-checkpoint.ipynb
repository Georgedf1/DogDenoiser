```python
import numpy as np
import tensorflow as tf
import sys, os


print("TensorFlow version: {}".format(tf.__version__))
print("Eager execution: {}".format(tf.executing_eagerly()))
```

```python
import bvh_utils
import marker_utils
import plot_utils
import neural_lbs_utils
```

```python
'''
NB: testSeq has different shape in marker data and joint data by 1, could easily be corrected
but for ease I will leave 'testSeq' out of motion_names below
'''

motion_names = ['jump','poles','table','trot','walk']

markers = []
joints = []

for name in motion_names:
    bvh_class = bvh_utils.BVHData()
    bvh_class.bvhRead('neural_lbs_data/dog1_' + name + '_skeleton.bvh')
    if joints == []:
        joints = bvh_class.getJointAngles()
        current_joints = joints
        markers = marker_utils.read_markers('neural_lbs_data/dog1_' + name + '_markers.json')
        current_markers = markers
    else:
        current_joints =bvh_class.getJointAngles()
        joints = np.append(joints, current_joints, axis=0)
        current_markers = marker_utils.read_markers('neural_lbs_data/dog1_' + name + '_markers.json')
        markers = np.append(markers, current_markers, axis=0)
    
    print('Shape equal: ',current_markers.shape[0]==current_joints.shape[0])
    if current_markers.shape[0] != current_joints.shape[0]:
        print('FALSE with joints: ',current_joints.shape[0],'and markers: ',current_markers.shape[0])

print('Finished')
print('Joint shape: ',joints.shape)
print('Marker shape: ',markers.shape)
```

```python
#Make neural lbs class
nn = neural_lbs_utils.Neural_LBS()

#Load the data
nn.load_tr_data(joints,markers)

print(nn.marker_data.shape)
print(nn.joint_data.shape)
```

```python
#Compile model
nn.compile()
```

```python
#Train model, using test data as validation for now
nn.train()
nn.train()
```

```python
bvh = bvh_utils.BVHData()
bvh.bvhRead('neural_lbs_data/dog1_testSeq_skeleton.bvh')
joints_testseq = bvh.getJointAngles()
markers_testseq = marker_utils.read_markers('neural_lbs_data/dog1_testSeq_markers.json')
```

```python
markers_testseq = np.delete(markers_testseq,0,0) # delete a frame to make equal shape
joints_local = bvh_utils.make_joints_local(joints_testseq)
local_tfs = bvh_utils.create_local_tfs_root(joints_testseq)
markers_local = marker_utils.make_markers_local(markers_testseq,local_tfs)

ani = plot_utils.AnimatePlot()
ani.animated_plot(joints_local[:,:,:,3],nn.predict(joints_local[:]))
```
