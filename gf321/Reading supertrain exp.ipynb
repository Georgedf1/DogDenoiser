```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
with open('supertrain_exp_1208.txt', 'r') as f:
    lines = f.readlines()
```

```python
lines[24286-4].split()
```

```python
results = [ [], [],[],[],[],[] ] 

for ind, line in enumerate(lines):
    
        split_line = line.split()
        
        if split_line[0] == 'Time':
            
            results[0].append(float(split_line[-1]))
            results[1].append(float(lines[ind-5].split()[-1]))
            try: #dealing with odd lines; only 2 of them
                results[2].append(float(lines[ind-4].split()[-1]))
                last_try = float(lines[ind-4].split()[-1])
            except:
                results[2].append(last_try) 
            results[3].append(float(lines[ind-3].split()[-1]))  
            results[4].append(float(lines[ind-2].split()[-1]))  
            results[5].append(float(lines[ind-1].split()[-1]))  
```

```python
results_np = np.array(results)
```

```python
results_np.shape
```

```python
fig = plt.figure()
ax = plt.axes()

ax.plot(results_np[0],results_np[1], label='tr')
ax.plot(results_np[0],results_np[2], label='cor_tr (training loss)')
ax.plot(results_np[0],results_np[3], label='val')
ax.plot(results_np[0],results_np[4], label='cor_val')
ax.plot(results_np[0],results_np[5], label='or_val')

plt.legend()
plt.show()
```
