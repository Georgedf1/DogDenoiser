```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from gf321_utils import bvh_utils
from gf321_utils import marker_utils
import numpy as np
```

```python
markers = np.load('neural_lbs_data/all_markers_aug.npy')
joints = np.load('neural_lbs_data/all_joints_aug.npy')
```

```python
pca_energy = 0.999
```

```python
markers_flat = markers.reshape((markers.shape[0],markers.shape[1]*3))

scaler = StandardScaler()
scaler.fit(markers_flat)

markers_n = scaler.transform(markers_flat)

pca = PCA(pca_energy)
pca.fit(markers_n)

markers_pca = pca.transform(markers_n)

print(markers.shape)
print(markers_pca.shape)
print('Compression ratio is: ', markers_pca.shape[1]/(markers.shape[1]*3))
```

```python
joints_flat = joints.reshape((joints.shape[0],joints.shape[1]*3*4))

scaler = StandardScaler()
scaler.fit(joints_flat)

joints_n = scaler.transform(joints_flat)

pca = PCA(pca_energy)
pca.fit(joints_n)

joints_pca = pca.transform(joints_n)

print(joints.shape)
print(joints_pca.shape)
print('Compression ratio is: ', joints_pca.shape[1]/(joints.shape[1]*4*3))
```
